# Muestra los países con esperanza de vida femenina mayor al promedio general y sus datos

```python
OCDE_vida=OCDE.set_index("INDICATOR")
expectativa = OCDE_vida.loc["HS_LEB"]
pd.DataFrame(expectativa)
```

```python
expectativa=expectativa.set_index('Inequality')
```

```python
mujeres_expectativa=expectativa.loc['Women']
```

```python
mujeres_expectativa.describe()
```

```python
valores_mujer = mujeres_expectativa['Value']
## Pregunta cuáles países tienen expectativa de vida mayor al promedio

#Parámetro de filtro = ['Value']restricción
promedio_vida=82.24
total=valores_mujer.count()

P_V = sum(valores_mujer>promedio_vida)
p_promedio = P_V/total

print("Existen ", P_V, " países con expectativa mayor al promedio", "de un total de ", total)
print("Esto representa el ", round(p_promedio,4)*100, "% del total de la muestra, según un parámetro promedio_vida= ",promedio_vida)
```

```python
#Para conocer cuáles países son los de mejor esperanza de vida (arriba del promedio)
#mujeres_expectativa=mujeres_expectativa.drop(["MEASURE", "INEQUALITY","Unit Code","PowerCode Code","Reference Period Code","Flag Codes","Flags"],axis=1)
Expectativa_arriba_media = mujeres_expectativa[mujeres_expectativa['Value'] > 82.24 ]
#Puedo acomodar el set de datos, de mayor a menor expectativa... 
Expectativa_arriba_sorteada=Expectativa_arriba_media.sort_values(by=['Value'],axis=0,ascending=False)
pd.DataFrame(Expectativa_arriba_sorteada)
```

# Parte 5. Información Gráfica

```python
## Esto es para extraer solamente los datos de la Inseguridad en los paises
DatosInseguridad = OCDE[OCDE['Indicator'] == 'Labour market insecurity']
DatosInseguridad = DatosInseguridad[DatosInseguridad['Inequality'] == 'Total']
```

```python
DatosInseguridad.drop(["MEASURE", "INEQUALITY","Unit Code","PowerCode Code","Reference Period Code","Flag Codes","Flags"],axis=1)
pd.DataFrame(DatosInseguridad)
```

```python
DatosInseguridad['Value'].plot(kind='line')
plt.xlabel('Países')
plt.ylabel('Nivel de inseguridad laboral')
plt.title('Porcentaje de Inseguridad Laboral en cada país')
plt.show()
```

```python
## Datos de la variable Calidad de Soporte a la educación

CalidadEducativa = OCDE[OCDE['Indicator'] == 'Years in education']
CalidadEducativa = CalidadEducativa[CalidadEducativa['Inequality'] == 'Total']
CalidadEducativa.drop(["MEASURE", "INEQUALITY","Unit Code","PowerCode Code","Reference Period Code","Flag Codes","Flags"],axis=1)
pd.DataFrame(CalidadEducativa)
```

```python
CalidadEducativa['Value'].plot(kind='line')
plt.xlabel('Países')
plt.ylabel('Calidad educativa')
plt.title('Porcentaje de Calidad educativa en cada país')
plt.show()
```
