### ¿En qué meses tengo que hacer combos de productos para elevar mi ticket promedio?

```python
## Importar las librerías
import pandas as pd  
import numpy as np   
import matplotlib.pyplot as plt  
```

```python
## importar el archivo
prediccion = pd.read_csv('predi2015python.csv')
```

```python
prediccion.head()
```

```python
prediccion.set_index('Indice')
```

```python
train = prediccion # entrenemos con todos los precions
```

```python
train.plot.scatter(x='Indice', y='PxN')
plt.show()
```

```python
plt.scatter(train.Indice,train.PxN, color='blue') # elegimos la latitud porque presenta mayor relacion lineal
a,b = np.polyfit(train.Indice, train.PxN, 1) # cálculo de los coeficientes ordenada al origen y pendiente 
                                                       # en una sola línea
x1 = min(train.Indice) # definamos los límites de la recta para dibujarla
x2 = max(train.Indice)
plt.plot([x1,x2], [a*x1 + b, a*x2 + b], color='red')
plt.show() 
print ("El valor de a y b es: ", a,b)
```

```python
#AQUI CHECO LA IMPORTANCIA DE LA VARIABLE, la longitud no importa tanto realmente 
# Correlation coefficients (r values)
cc = np.corrcoef(train.Indice, train.PxN)[1,0]
print('Correlation coefficient for precio versus demanda:', cc)
cc = np.corrcoef(train.Indice, train.PxN)[1,0]
print('Correlation coefficient for precio versus demanda:', cc)
```
